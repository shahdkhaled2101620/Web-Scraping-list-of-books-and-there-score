```python
import requests
from bs4 import BeautifulSoup
```

```python
link='https://www.goodreads.com/list/show/153860.Goodreads_Top_100_Highest_Rated_Books_on_Goodreads_with_at_least_10_000_Ratings'
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}


page = requests.get(link, headers=headers)
```

```python
soup = BeautifulSoup(page.text, 'html.parser') #get html code 
```

```python
book = soup.find_all('tr', itemtype="http://schema.org/Book")
```

```python
for i in book:
    btitle=i.find('span',{'itemprop':"name"})
    auther=i.find('span',{'itemprop':"author"})
    score=i.find('a',{"href":"#"})
    #print(score.text.strip())
```

```python
import pandas as pd
```

```python
data=[]
```

```python
for i in book:
    btitle=i.find('span',{'itemprop':"name"})
    auther=i.find('span',{'itemprop':"author"})
    score=i.find('a',{"href":"#"})

    data.append({
        #علشان لو عنصر مش موجود → البرنامج ما يقعش
        "Tilte": btitle.text.strip() if btitle else None,
        "Auther": auther.text.strip() if auther else None,
        "Score": score.text.strip() if score else None
    })
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_excel(r'practis_webscraping.ipynb2.xlsx')
```

```python
df1 = pd.read_excel("practis_webscraping.ipynb.xlsx")
df2 = pd.read_excel("practis_webscraping.ipynb2.xlsx")
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)
```

```python
df_merged.to_excel("books_merged.xlsx", index=False)
```

```python
df = pd.read_excel("books_merged.xlsx")
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df["Score"] = (
    df["Score"]
    .str.replace("score:", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
    .astype(int)
)
```

```python
df = df.sort_values(by="Score", ascending=False)
```

```python
df = df.reset_index(drop=True)
```

```python
df.insert(0, "Rank", df.index + 1)
```

```python
df.to_excel("books_clean.xlsx", index=False)
```
